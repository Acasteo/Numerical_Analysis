```python
import numpy as np
import matplotlib.pyplot as plt
def tri_diag(A,d):
    m_n = np.array([A[0][0]])
    l_n = np.array([])
    b_n = np.array([])
    for j in range(1, len(A[0])):
        c_j = A[j][j-1]
        a_j = A[j][j]
        b_j = A[j-1][j]
        l_j = c_j/m_n[j-1]
        m_j = a_j - l_j*b_j
        b_n = np.append(b_n, b_j)
        l_n = np.append(l_n, l_j)
        m_n = np.append(m_n, m_j)
    y_n = np.array([d[0]])
    for j in range(1,len(A)):
        y_j = d[j] - l_n[j-1]*y_n[j-1]
        y_n = np.append(y_n, y_j)
    x_n = np.array([y_n[-1]/m_n[-1]])
    for j in range(2, len(A[0])):
        x_j = (y_n[-j]- b_n[-j]*x_n[0])/m_n[-j]
        x_n = np.insert(x_n, 0, x_j)
    x_j = (y_n[0]- b_n[0]*x_n[0])/m_n[0]
    x_n = np.insert(x_n, 0, x_j)
    return x_n
```

```python
A = [[1,1,0,0,0],[1,2,1,0,0],[0,1,3,1,0],[0,0,1,4,1],[0,0,0,1,5]]
b = [1*9,2*7,3*6,4*3,5*4]
```

#### (b)

```python
tri_diag(A,b)
```

```python
np.linalg.solve(A,b)
```

```python
def rhs(x):
    return 2*(np.pi**2)*np.sin(np.pi*x)
def c_func(x):
    return np.pi**2*(x+1)/(x+1)
```

```python
N = 50
h = 1/N
x = np.linspace(0,1, N+1)
x = x[1:-1]

A = np.array([np.zeros_like(x)])
A[0][0] = 2/(h**2) + c_func(x[0])
A[0][1] = -1/(h**2)
for i in range(1,len(x)-1):
    A = np.insert(A,i, [np.zeros_like(x)],axis = 0)
    A[i][i-1] = -1/(h**2)
    A[i][i] = 2/(h**2) + c_func(x[i])
    A[i][i+1] = -1/(h**2)
A = np.insert(A,i+1, [np.zeros_like(x)],axis = 0)
A[-1][-1] = 2/(h**2) + c_func(x[-1])
A[-1][-2] = -1/(h**2)
d = rhs(x)

solve = tri_diag(A,d)
solve = np.append(solve, 0)
solve = np.insert(solve, 0,0)
x = np.insert(x,0,0)
x = np.append(x,1)
```

```python
actual = np.sin(np.pi*x)
plt.figure(figsize=(30,15))
plt.plot(x,solve, '-g')
plt.plot(x,actual)
plt.grid()
plt.plot()
plt.show()
```

$f(x) = sin(\pi x) $

eq 1) $-u'' + \pi^2 u = 2\pi^2 sin(\pi x)$

Consider the LHS of eq1 with $u(x) = f(x)$

$-\frac{d^2}{dx^2} sin(\pi x) + \pi^2 sin(\pi x) $


$ \iff -\frac{d}{dx} \pi cos(\pi x) + \pi^2 sin(\pi x) $

$ \iff \pi^2 sin(\pi x) + \pi^2 sin(\pi x) = 2 \pi^2 sin(\pi x) $

which is exactly the RHS as desired
