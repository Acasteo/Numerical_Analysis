```python
import numpy as np

# This is an implementation of Newton's divided difference coefficents
# Written by Austin Castelo
# 11/8/2018
# INPUT: list of values x and list of their output of the function f
# OUPUT: array of divided difference coefficents
def c_j(f,x):
    c_j = np.zeros(len(x))
    for j in range(len(x)):
        c_j[j] = f[j]
    for k in range(1, len(x)):
        for j in range(1,len(x)-k+1):
            c_j[-j] = (c_j[-j] - c_j[-(j+1)])/(x[-j] - x[-(j+k)])
        
    return c_j

# This is an implementation of Newton's interpolating polynomial using Horner-like scheme
# Written by Austin Castelo
# 11/8/2018
# INPUT: list of values x and list of their output of the function f, value to approx x, and list of divided 
# difference coefficents c_j
# OUPUT: p_n approximation of f(x)
def p(f,x_n, x, c_j):
    c = c_j
    p = c[-1]
    for j in range(2, len(c)+1):
        p = c[-j]+(x-x_n[-j])*p
    return p
def f(x):
    return x**2
```

```python
x_n = np.linspace(0,3,4)
```

```python
f = f(x_n)
cⱼ = c_j(f,x_n)
cⱼ
```

$c_0 = 0$

$c_1 = 0 + \frac{1}{1}=1$

$c_2 = 0 + \frac{1}{(-1)(1)} + \frac{4}{(2)(1)} = 1$

$c_3 = 0 + \frac{1}{(-2)(-1)(1)} + \frac{4}{(2)(1)(-1)} + \frac{9}{(3)(2)(1)} = 0$

Therefore the $c_j$ function is accurate.


```python
p(f,x_n,5, cⱼ)
```

$f(5)=25$

$p_n(f,x_n,5) = 25$

Therefore $p_n$ function is also accurate.

```python
def f(x):
    return np.e**(-x**2)

# This is an implementation of interpolating nodes
# Written by Austin Castelo
# 11/9/2018
# INPUT: number of nodes j
# OUPUT: array of length j+1 of nodes on [-1,1]
def x_j(j):
    x = np.array([])
    for i in range(j+1):
        x_j = -1 + i*(2/j)
        x = np.append(x, x_j)
    return x
xⱼ = x_j(10)
fⱼ = f(xⱼ)
cⱼ = c_j(fⱼ,xⱼ)
x̄ⱼ = x_j(100)
fₓ = f(x̄ⱼ)
pₓ = np.array([])
```

```python
for x in x̄ⱼ:
    pₓ = np.append(pₓ, p(fⱼ, xⱼ, x, cⱼ))
```

```python
error = fₓ - pₓ
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,15), facecolor= "w")
plt.plot(x̄ⱼ, error
        )
plt.legend(labels = ('Error',), fontsize = 20)
plt.grid()
plt.show()
```

```python
# This is an implementation of coefficents for Hermite interpolation
# Written by Austin Castelo
# 11/9/2018
# INPUT: list of values x and list of their output of the function f
# OUPUT: array of Hermite interpolation coefficents
def c_j(f,x):
    c_j = np.zeros(len(x))
    for j in range(len(x)):
        c_j[j] = f[j]
    for k in range(1, len(x)):
        for j in range(1,len(x)-k+1):
            if (c_j[-j] - c_j[-(j+1)]) == 0 and x[-j]== 0:
                c_j[-j] = 0
                break
            elif (c_j[-j] - c_j[-(j+1)]) == 0 and x[-j]== 1:
                c_j[-j] = 3
                break
            c_j[-j] = (c_j[-j] - c_j[-(j+1)])/(x[-j] - x[-(j+k)])
        
    return c_j
x = np.array([0,0,1,1])
f = [0,0,2,2]
c = c_j(f,x)
c
```

$p(x)=(2)(x^2)+ (-1)(x^2)(x-1) = 2x^2 - (x^3-x^2) = 3x^2-x^3$

```python
p(f,x,2,c)
```
