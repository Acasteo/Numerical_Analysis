```python
import numpy as np
import matplotlib.pyplot as plt

# This is an implementation of a Chebyshev polynomial of degree n
# Written by Austin Castelo
# 10/19/2018
# INPUT: Degree pf polynomial n and point x
# OUPUT: the value of the Chebyshev at point x
def T_n(n,x):
    if n == 1:
        return x
    elif n == 0:
        return 1
    return 2*x*T_n(n-1,x)-T_n(n-2,x)
```

```python
x = np.linspace(0,1, 1000)
y = []
for j in [1,2,3,6]:
    y_j = []
    for i in x:
        y_j = np.append(y_j, T_n(j,i))
    y.append(y_j)
```

```python
plt.figure(figsize=(30,15), facecolor= "w")
plt.plot(x, y[0],
         x, y[1],
         x, y[2],
         x, y[3],
         
        )
plt.legend(labels = ('$T_1$','$T_2$','$T_3$','$T_6$'), fontsize = 20)
plt.grid()
plt.show()
```
