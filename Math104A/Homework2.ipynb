```python
import numpy as np
import scipy.special

# This is an implementation of Bernstein Polynomials
# Written by Austin Castelo
# 10/12/2018
# INPUT: number of nodes n, function f, and point x to approximate f(x)
# OUPUT: approximation of f(x)
def B_P(n,f,x):
    total = 0
    for k in range(0,n):
        total += f(k/n)*scipy.special.comb(n,k)*(x**k)*((1-x)**(n-k))
    return total
def f(x):
    if 0 <= x <= 0.5:
        return x
    elif 0.5 < x <=1:
        return 1-x
# This is an implementation of the error |f(x)-B_P| on an interval
# Written by Austin Castelo
# 10/12/2018
# INPUT: number of nodes n, function f, and array of points x to approximate f(x)
# OUPUT: an x length array of errors
def E_n(n,f,x):
    E = []
    for i in x:
        E = np.append(E,np.abs(f(i)-B_P(n,f,i)))
    return E
```

```python
x = np.linspace(0,1, 500)
E_list = [[0]]
max_En = []
for n in range(1,50):
    E_list.append([E_n(n,f,x)])
    print(np.max(E_list[n]))
    max_En.append(np.max(E_list[n]))
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,n-1), (np.asarray(max_En[0:n-2])/np.asarray(max_En[1:n-1])))
```

```python
np.asarray(max_En[0:n-2])/np.asarray(max_En[1:n-1])
```

The ratio of $||\bullet ||_{\infty}$ of the nth and nth+1 errors as $n \rightarrow \infty$ approachs 1. Thus the rate of convergence of $B_n(f(x))$ to $f(x)$ is 1 and the convergence is sublinear.

```python
x = np.linspace(0,1,50)
plt.figure(figsize=(30,15), facecolor= "w")
plt.plot(x, B_P(50,f,x),
         x, B_P(30,f,x),
         x, B_P(10,f,x),
         x, B_P(4,f,x),
         x, B_P(2,f,x),
         x, B_P(1,f,x),
         [0,0.5,1], [0,0.5,0],
         
        )
plt.legend(labels = ('50','30','10','4','2','1',"f(x)"), fontsize = 20)
plt.grid()
plt.show()
```

```python
# This is an implementation of Bezier Curve for x cordinates
# Written by Austin Castelo
# 10/12/2018
# INPUT: three points p0,p1,p2
# OUPUT: array of x-cordinates of Bezier Curve with 500 elements
def Bez_x(p0,p1,p2):
    t = np.linspace(0,1,500)
    x = []
    for t in t:
        B2_t = (1-t)**2*p0 + 2*t*(1-t)*p1 + t**2*p2
        x = np.append(x, B2_t)
    return x

# This is an implementation of Bezier Curve for y cordinates
# Written by Austin Castelo
# 10/12/2018
# INPUT: three points p0,p1,p2
# OUPUT: array of y-cordinates of Bezier Curve with 500 elements
def Bez_y(p0,p1,p2):
    t = np.linspace(0,1,500)
    y = []
    for t in t:
        B2_t = (1-t)**2*p0 + 2*t*(1-t)*p1 + t**2*p2
        y = np.append(y, B2_t)
    return y
```

```python
x = np.append(Bez_x(12,11.5,11), Bez_x(11,10.5,10))
y = np.append(Bez_y(20,25,10), Bez_y(10,0,5))
```

```python
plt.figure(figsize=(30,15), facecolor= "w")
plt.ylim(-5,30)
plt.xlim(0,30)
plt.plot(x,y)
plt.grid()
plt.show()
```
