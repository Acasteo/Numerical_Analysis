```python
import numpy as np
import matplotlib.pyplot as plt

def RK4(yn,xn,h,xp, yp):
    K1 = h*yp(xn,yn)
    C1 = h*xp(xn,yn)
    K2 = h*yp(xn+C1/2, yn+K1/2)
    C2 = h*xp(xn+C1/2, yn+K1/2)
    K3 = h*yp(xn+C2/2, yn+K2/2)
    C3 = h*xp(xn+C2/2, yn+K2/2)
    K4 = h*yp(xn+C3, yn+K3)
    C4 = h*xp(xn+C3, yn+K3)
    yn1 = yn+1/6*(K1+2*K2+ 2*K3+K4)
    xn1 = xn+1/6*(C1+2*C2+ 2*C3+C4)
    return yn1, xn1

def xp(x,y):
    return 1.2*x-x**2-(x*y)/(x+0.2)

def yp(x,y):
    return (1.5*x*y)/(x+0.2)-y    
```

```python
def graph(y0,t0,x0,h,T):
    y = np.array([y0])
    x = np.array([x0])
    t = np.array([t0])
    n = 0
    while(t0+h*n < T):
        yn, xn = RK4(y[n], x[n], h, xp, yp)
        y = np.append(y, yn)
        x = np.append(x, xn)
        t = np.append(t, t0+h*n)
        n += 1
    plt.figure(figsize=(13,8) , facecolor= 'w')
    plt.plot(t, y, '-b', linewidth = 3)
    plt.plot(t, x, '-r', linewidth = 3)
    plt.grid(linewidth = 2)
    plt.legend(labels = ['Predator', 'Prey'])
    plt.plot()
    plt.show()
```

```python
graph(0.75,0,1,0.1,30)
```

```python
graph(0.25,0,0.75,0.1,30)
```

The results of the model with both sets of initial conditions show a cyclical relationship between predator and prey where as predators increase prey decreases and as prey increases so do predators until they force a decrease in growth of prey.
