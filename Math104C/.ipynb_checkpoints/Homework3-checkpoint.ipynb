```python
import numpy as np
import matplotlib.pyplot as plt

def u0x(x):
    for i in range(len(x)):
        if x[i] > np.pi/2:
            x[i] = np.pi-x[i]
    return x

def fin_dif(M,N,sig,T):
    k = T/M
    h = np.pi/N
    alpha = sig*k/(h**2)
    print(alpha)
    x = np.linspace(0,np.pi, N)
    u0 = np.array([u0x(x)])
    u = np.repeat(u0, M, axis=0)
    for n in range(1,M-1):
        for j in range(1,N-1):
            u[n+1][j] = u[n][j] + alpha*(u[n][j-1]-2*u[n][j]+u[n][j+1])
    return u
```

```python
u = fin_dif(1000,20,1,10)
```

```python
def graph(M,N,sig,T):
    u = fin_dif(M,N,sig,T)
    x = np.linspace(0,np.pi, N)
    plt.figure(figsize=(13,8) , facecolor= 'w')
    for i in range(len(u)):
        plt.plot(x, u[i], color = np.random.rand(4), linewidth = 1.5)
    plt.grid(linewidth = 2)
    plt.show()
    for i in range(1, 10):
        print(np.sum(abs(u[-(i+1)]-u[-(i+2)]))/np.sum(abs(u[-i]-u[-(i+1)])))
```

```python
graph(2000,90,1,1)
```

```python
graph(20,11,1,1)
```

```python
def tri_diag(A,d):
    m_n = np.array([A[0][0]])
    l_n = np.array([])
    b_n = np.array([])
    for j in range(1, len(A[0])):
        c_j = A[j][j-1]
        a_j = A[j][j]
        b_j = A[j-1][j]
        l_j = c_j/m_n[j-1]
        m_j = a_j - l_j*b_j
        b_n = np.append(b_n, b_j)
        l_n = np.append(l_n, l_j)
        m_n = np.append(m_n, m_j)
    y_n = np.array([d[0]])
    for j in range(1,len(A)):
        y_j = d[j] - l_n[j-1]*y_n[j-1]
        y_n = np.append(y_n, y_j)
    x_n = np.array([y_n[-1]/m_n[-1]])
    for j in range(2, len(A[0])):
        x_j = (y_n[-j]- b_n[-j+1]*x_n[0])/m_n[-j]
        x_n = np.insert(x_n, 0, x_j)
    x_j = (y_n[0]- b_n[0]*x_n[0])/m_n[0]
    x_n = np.insert(x_n, 0, x_j)
    return x_n
```

```python
def back_def(M,N,sig,T):
    k = T/M
    h = np.pi/N
    alpha = sig*k/(h**2)
    print(alpha)
    x = np.linspace(0,np.pi, N)
    u0 = np.array([u0x(x)])
    u = np.repeat(u0, M, axis=0)
    A = np.zeros((M,N))
    A[0][0] = 1
    A[-1][-1] = 1
    for i in range(1, N-1):
        A[i][i-1] = -1*alpha
        A[i][i] = 1+2*alpha
        A[i][i+1] = -1*alpha
    for i in range(1,N):
        u[i] = tri_diag(A,u[i-1])
    return u
```

```python
def graph2(M,N,sig,T):
    u = back_def(M,N,sig,T)
    x = np.linspace(0,np.pi, N)
    plt.figure(figsize=(13,8) , facecolor= 'w')
    for i in range(len(u)):
        plt.plot(x, u[i], '-b', linewidth = 0.5)
    plt.grid(linewidth = 2)
    plt.show()
    for i in range(1, 10):
        print(np.sum(abs(u[-(i+1)]-u[-(i+2)]))/np.sum(abs(u[-i]-u[-(i+1)])))
```

```python
graph2(1000,1000,1,np.pi)
```
