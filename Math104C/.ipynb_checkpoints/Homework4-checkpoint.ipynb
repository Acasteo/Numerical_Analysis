```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
def tri_diag(A,d):
    m_n = np.array([A[0][0]])
    l_n = np.array([])
    b_n = np.array([])
    for j in range(1, len(A[0])):
        c_j = A[j][j-1]
        a_j = A[j][j]
        b_j = A[j-1][j]
        l_j = c_j/m_n[j-1]
        m_j = a_j - l_j*b_j
        b_n = np.append(b_n, b_j)
        l_n = np.append(l_n, l_j)
        m_n = np.append(m_n, m_j)
    y_n = np.array([d[0]])
    for j in range(1,len(A)):
        y_j = d[j] - l_n[j-1]*y_n[j-1]
        y_n = np.append(y_n, y_j)
    x_n = np.array([y_n[-1]/m_n[-1]])
    for j in range(2, len(A[0])):
        x_j = (y_n[-j]- b_n[-j+1]*x_n[0])/m_n[-j]
        x_n = np.insert(x_n, 0, x_j)
    x_j = (y_n[0]- b_n[0]*x_n[0])/m_n[0]
    x_n = np.insert(x_n, 0, x_j)
    return x_n
```

```python
def ADI(M,N,sig,T):
    k = T/M
    h = 1/N
    alpha = sig*k/(2*h**2)
    x = np.linspace(0, 1, N)
    y = np.linspace(0, 1, N)
    u0 = np.zeros((N,N))
    A1 = np.zeros((N,N))
    A2 = np.zeros((N,N))
    A1[0][0] =  A2[0][0] = 1
    A1[-1][-1] =  A2[-1][-1] = 1
    for i in range(N-1):
        for j in range(N-1):
            u0[i][j] = np.sin(np.pi*x[i])*np.sin(np.pi*y[j])
    u = np.zeros((M,N,N))
    u[0] = u0
    u_s = np.zeros((M,N,N))
    for i in range(1,N-1):
        A1[i][i-1] = -1*alpha
        A1[i][i] = 1+2*alpha
        A1[i][i+1] = -1*alpha
        
        A2[i][i-1] = 1*alpha
        A2[i][i] = 1-2*alpha
        A2[i][i+1] = 1*alpha
    for n in range(M-1):
        for i in range(N):
            d = np.matmul(A2,u[n][i])
            u_s[n][i] = tri_diag(A1,d)
            d = np.matmul(A2,u_s[n][i])
            u[n+1][i] = tri_diag(A1,d)
    return u
```

The ADI method is unconditionally stable, thus numerical stability does not play a role in our choice of k and h. Thus we take k and h to be sufficently small for the resolution we wish to achieve.

## b)

```python
u1 = ADI(50,50, 1, 1)
u2 = ADI(50,100, 1, 1)
u3 = ADI(50,200, 1, 1)
```

```python
print((np.sum(u3[-1])-np.sum(u2[-1]))/(np.sum(u2[-1])-np.sum(u1[-1])))
```

Thus we can see the method has approximately $O(h^2)$ convergence rate.

For this particular problem I would use the ADI method over Crank-Nicolson. This is because the ADI method allows us to form tridiagonal systems of equations to solve while Crank-Nicolson does not, and the efficency and easy of solving tridiagonal systems makes the ADI method the superior choice.
