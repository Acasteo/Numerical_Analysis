```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u1(x):
    return np.piecewise(x, [abs(x)<=1, abs(x)>1], [lambda x: 1-abs(x), lambda x: 0])
def u2(x):
    return np.e**(-5*x**2)
```

```python
def ftfs(h, l, u):
    k = l*h
    M = int(4/h)
    x = np.linspace(-1,3, M)
    v0 = u(x)
    v0[-1] = v0[-2]
    v = np.zeros((40,M))
    v[0] = v0
    for n in range(39):
        for j in range(M-1):
            v[n+1][j] = (1+l)*v[n][j]-l*v[n][j+1]
        v[n+1][-1] = v[n+1][-2]
    return v, x
```

```python
def graph(h, l, u):
    v,x = ftfs(h,l, u)
    for i in [0,10,25,38]:
        plt.figure(figsize=(13,8) , facecolor= 'w')
        plt.plot(x, v[i], color = 'b', linewidth = 1.5)
        plt.grid(linewidth = 2)
        plt.show()
```

```python
graph(0.02, 0.8, u1)
```

```python
graph(0.02,0.8,u2)
```

As we can see for the first initial condition we see instability by time step 10, while for the second initial condition the graph still appears stable at time step 25. Thus, we see instability appears sooner for the less smooth initial condition.

Stability condition of forward space and forward time scheme is $|a| \frac{k}{h} \leq 1$.

$a = 1 \rightarrow |1|0.8= 0.8 \leq 1$

Thus, the scheme is stable.

```python
def ft1fs2(h, l, u):
    k = l*h
    alpha = l/h*0.01
    M = int(4/h)
    x = np.linspace(-1,3, M)
    v0 = u(x)
    v0[0] = 0
    v = np.zeros((40,M))
    v[0] = v0
    for n in range(39):
        for j in range(1,M):
            v[n+1][j] = (1-l)*v[n][j]+l*v[n][j-1]
    return v, x
```

```python
def graph2(h, l, u):
    v,x = ft1fs2(h,l, u)
    for i in [0,10,25,38]:
        plt.figure(figsize=(13,8) , facecolor= 'w')
        plt.plot(x, v[i], color = 'b', linewidth = 1.5)
        plt.grid(linewidth = 2)
        plt.show()
```
