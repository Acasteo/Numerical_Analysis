```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t,y):
    return (3*np.cos(t))/(y**2)
def Euler(N,y0,t):
    k = np.pi/N
    print(k)
    y = np.array([y0])
    for i in range(N-1):
        yⱼ = y[i] + k*(f(t[i], y[i]))
        y = np.append(y, yⱼ)
    return y
def sol(t):
    return (9*np.sin(t)+1)**(1/3)
```

```python
def graph(N,y0):
    t = np.linspace(0,np.pi, N)
    sol1 = sol(t)
    approx = Euler(N,y0,t)
    plt.figure(figsize=(13,8) , facecolor= 'w')
    plt.plot(t,sol1, '-b', linewidth = 3)
    plt.plot(t,approx, '-r', linewidth = 3)
    plt.grid(linewidth = 2)
    plt.legend(labels = ['Solution', 'Euler Method Approximation'])
    plt.plot()
    plt.show()
    error = 0
    print('Error = ' + str(sum(abs(approx-sol1)**2)))
    
```

```python
graph(50,1)
```

```python
graph(100,1)
```

```python
graph(200,1)
```

Our expected rate of convergence is $O(k)$, which can be verified using the convergence of the error. With $k = 0.031415926535897934
$ the error is $0.19427481668649027$, and with $k = 0.015707963267948967$ the error is $0.10125782214841086$. 

If the convergence is $O(k)$ we would expect the error to decrease by $\frac{0.015707963267948967}{0.031415926535897934} = 0.5$.

The error decreased by $\frac{0.10125782214841086}{0.19427481668649027} = 0.521209201868994766 \approx 0.5$. Thus we have verified the rate of convergence is $O(k)$ for sufficently small k.

### 4 b)

$\alpha = 0.2$

$\omega = 2$

$\theta(0) = 1$

$\theta^{'}(0)=0$

$\theta^{''}+\alpha \theta^{'}+ \omega^{2}\theta=0$

System of equations

$\theta_1 = \theta \rightarrow \theta_1^{'}= \theta_2$

$\theta_2 = \theta^{'} \rightarrow \theta_2^{'} = -\alpha\theta_2-\omega^2\theta_1$


```python
def θ_p(θ1,θ2):
    return -0.2*θ2 - 4*θ1
def Euler_sys(N,θ1_0,θ2_0,T):
    k = T/N
    print(k)
    θ1 = np.array([θ1_0])
    θ2 = np.array([θ2_0])
    for i in range(N-1):
        θ1ⱼ = θ1[i]+k*θ2[i]
        θ2ⱼ = θ2[i]+k*θ_p(θ1[i],θ2[i])
        θ1 = np.append(θ1, θ1ⱼ)
        θ2 = np.append(θ2, θ2ⱼ)
    return θ1
def sol(t):
    return np.e**(-0.1*t)*(np.cos((np.sqrt(399))*t/10)+0.1/1.9975*np.sin((np.sqrt(399))*t/10))
```

```python
def graph(N,θ1_0,θ2_0,T):
    t = np.linspace(0,T, N)
    sol1 = sol(t)
    approx = Euler_sys(N,θ1_0,θ2_0,T)
    plt.figure(figsize=(13,8) , facecolor= 'w')
    plt.plot(t,sol1, '-b', linewidth = 3)
    plt.plot(t,approx, '-r', linewidth = 3)
    plt.grid(linewidth = 2)
    plt.legend(labels = ['Solution', 'Euler Method Approximation'])
    plt.plot()
    plt.show()
    error = 0
    print('Error = ' + str(sum(abs(approx-sol1)**2)))
```

```python
graph(250,1,0,5)
```

```python
graph(500,1,0,5)
```

```python
graph(1000,1,0,5)
```

```python
graph(2000,1,0,5)
```

Our expected rate of convergence is $O(k)$, which can be verified using the convergence of the error. With $k = 0.005
$ the error is $0.23196276126793977$, and with $k = 0.0025$ the error is $0.11379576745071844$. 

If the convergence is $O(k)$ we would expect the error to decrease by $\frac{0.0025}{0.005} = 0.5$.

The error decreased by $\frac{0.11379576745071844}{0.23196276126793977} = 0.49057774113696273 \approx 0.5$. Thus we have verified the rate of convergence is $O(k)$ for sufficently small k.
